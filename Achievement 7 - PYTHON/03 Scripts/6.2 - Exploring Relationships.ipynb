# Task 3 - Correlation Heatmap

```python
# Creating a subplot using matplotlib library

f,ax = plt.subplots(figsize=(10,10))
```

```python
# Create the correlation heatmap in seaborn by applying a heatmap onto the 
# correlation matrix and the subplots defined above.

corr = sns.heatmap(sub.corr(), annot = True, ax = ax)
```

```python
corr.figure
```

# Task 4 - ScatterPlot

```python
#### Creating a scatterplot for the variables 'daily_vaccs' and 'total_vaccs'.
#### This is because these two variables have the strongest correlation with a coeffiencit of 0.93.

sns.lmplot(x = 'total_vaccs', y = 'daily_vaccs', data = sub)
```

#### The above scatterplot graph shows a positive trend line which depicts the positive relationship between the 2 variables 'Total Vaccinations' and 'Daily Vaccinations'.
#### This implies that when the total number of vaccinations are increased, the number of daily vaccinations will also increase.
